```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt


# 加载图片
im = Image.open('./../image/lena.png')
print(im.size)
```

```python
img = np.array(im)      # image类 转 numpy
print(img.shape)
```

```python
#
img = img[:,:,0:3]
plt.imshow(img, 'Blues')
plt.show()
```

```python
# 变大Magnification 倍得缩放矩阵

Magnification = 0.3

A= [
    [Magnification, 0, 0],
    [0, Magnification, 0],
    [0, 0, 1],
]
A = np.array(A)  
AI = np.linalg.inv(A)


x0 = [1,2,1]
x0 = np.array(x0).reshape(3,1)


```

```python
xNew = np.dot(A, x0)
xNew
```

```python
w,h,c = img.shape
newImg = np.ones((w, h,c))
# 临近插值算法
for i in range(int(w * Magnification)):
    for j in range(int(h * Magnification)):
        pointNow = np.array([i, j,1])
        pointNew = np.dot(AI, pointNow)
        srcX = round(pointNew[0])
        srcY = round(pointNew[1])
        newImg[i][j] = img[srcX][srcY]


newImg = newImg.astype(np.int32)        
plt.imshow(newImg)
# plt.show(0)
        
```

```python
# 旋转矩阵计算
theta = np.pi / 6

A= [
    [np.cos(theta), np.sin(theta), 0],
    [-1*np.sin(theta), np.cos(theta), 0],
    [0, 0, 1],
]
A = np.array(A)
AI = np.linalg.inv(A)
x0 = [1,2,1]
x0 = np.array(x0).reshape(3,1)


w, h, c = img.shape
newImg = np.zeros((w, h, c))


# 双线性插值插值算法
def linearInterpolation(x0, y0, x1,y1, xn):
    # 原来的256有限制数位的大小，会导致数据溢出
    y0= y0.astype(np.int32)
    y1 = y1.astype(np.int32)
    # 要小心被除数为0
    if x1 -x0 !=0:
        a =  y0 + (y1- y0) * ((xn- x0)/(x1-x0))

        return a
    else:
        return y0




# 双线性插值插值算法
for i in range(int(w )):
    for j in range(int(h)):
        try:


            pointNow = np.array([i, j, 1])
            pointNew = np.dot(AI, pointNow)
            srcX = pointNew[0]
            srcY = pointNew[1]

            # 剔除边界外的点
            if srcX >0 and srcX < w and srcY >0 and srcY < h:
                pass
            else:
                continue

            srcXMin = np.floor(pointNew[0]).astype(np.int32)
            srcYMin = np.floor(pointNew[1]).astype(np.int32)
            srcXMax = np.ceil(pointNew[0]).astype(np.int32)
            srcYMax = np.ceil(pointNew[1]).astype(np.int32)

            yUP = linearInterpolation(srcXMin, img[srcXMin][srcYMax], srcXMax, img[srcXMax][srcYMax], srcX)
            yDOWN = linearInterpolation(srcXMin, img[srcXMin][srcYMin], srcXMax, img[srcXMax][srcYMin], srcX)
            # print(srcX,srcY)
            ans = linearInterpolation(srcYMin, yDOWN, srcYMax, yUP, srcY)
            # print(str(ans))


            newImg[i][j] = ans
        except Exception as e:
            newImg[i][j] = [0,0,0]
            # print(e)



newImg = newImg.astype(np.int32)
# print(newImg)
plt.imshow(newImg)
plt.show()
```
