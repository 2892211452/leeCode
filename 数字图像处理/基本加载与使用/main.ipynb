```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt


# 加载图片
im = Image.open('./../image/lena.png')
print(im.size)
```

```python
img = np.array(im)      # image类 转 numpy
print(img.shape)
```

```python
#
img = img[:,:,0:3]
plt.imshow(img, 'Blues')
plt.show()
```
